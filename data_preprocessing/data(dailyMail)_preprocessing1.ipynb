# DailyMail 1차 데이터 전처리(21만 ->12만)

```python
import unicodedata
import re
import numpy as np
import pandas as pd
from os import listdir


# load all stories in a directory
def load_stories(directory):
    print('stories loading')
    stories = list()
    i =1
    for name in listdir(directory):
        filename = directory + '/' + name
        print(i)
        i = i+1
        #print(filename)   #-> C:/Users/Desktop/dataset/cnn_stories/cnn/stories/0001d1afc246a7964130f43ae940af6bc6c57f01.story
        
        # load document
        doc = load_doc(filename)
        
        # story 전처리 준비
        index = doc.find('By',0,10)  
        index1 = doc.find('PUBLISHED:',0,100)
        index2 = doc.find('UPDATED:',0,100)
        index3 = doc.find('@highlight')  
        story_string =""
        highlights_string=""
        
        
        #By, PUBLIEHD:, UPDATED: 없는 기사만 csv파일에 저장하기!
        if (index == -1) and (index1 == -1) and (index2 == -1): 
            # split into story and highlights
            story, highlights = doc[:index3], doc[index3:].split('@highlight')
            
            # story 전처리!
            story = [s for s in story.splitlines() if len(s) > 0]
            story_string = ' '.join(story)

            #highlights 전처리!
            highlights = [h.strip() for h in highlights if len(h) > 0]   
            # array형식으로 나눠져있는 문장데이터를 하나의 string으로 바꿈!
            highlights_string = '. '.join(highlights)
            highlights_string = highlights_string + '.'
            
            stories.append({'story':story_string, 'highlights':highlights_string})
            print('story - ', story_string[:50],'\n')
            #print('highlights - ', highlights_string,'\n')
        else :   
            pass
        
    return stories


# load doc into memory
def load_doc(filename):
    # open the file as read only
    file = open(filename, encoding='utf-8')
    # read all text
    text = file.read()
    # close the file
    file.close()
    return text
 
    
directory = 'C:/Users/Jeyoung/Desktop/dataset/dailymail_stories/dailymail/stories'
stories = load_stories(directory)
print('Loaded Stories %d' % len(stories))


df = pd.DataFrame(data = stories)
print(df.describe())
df.to_csv('dailymail_stories_preprocessing1.csv', index=False)
```
